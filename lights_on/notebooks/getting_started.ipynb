### Plotting Lights
For example, we can look at the nightlight satellite images of Barcelona from 1992 to 1997 as follows:

```python
plot_light_series('Barcelona_n41.39_e2.15', '1992-1997')
```

In the same fashion, we can also look at what is perhaps more interesting, the light changes between the years. To get e.g. the absolute change from year to year, be can use...

```python
plot_light_change('Barcelona_n41.39_e2.15', '1992-1997', style='RdBu')
```

We can also look at the growth instead of the absolute change and over different periods of time as below...

```python
plot_light_growth('Barcelona_n41.39_e2.15', '1992_1993_2012_2013', style='RdBu')
```

#### Styles
Whenever plotting the satellite data as an image, we can choose a style. The standard for displaying actual lights is usually `bone` and 'RdBu' for showing light differences, but there are other options. You can find the complete list of styles available through Matplotlib [here](https://matplotlib.org/examples/color/colormaps_reference.html).

```python
plot_light_series('Barcelona_n41.39_e2.15', '1993_2013', style='cubehelix')
```
