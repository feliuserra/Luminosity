### Plotting Lights
There are different functions for different plots of the data but two arguements are always required in order to plot a series, the `coordinates` (possibly including the location name) and the `dates` that we wish to explore.

* **Note: coordinate values should be entered in [decimal format](https://en.wikipedia.org/wiki/Decimal_degrees))**

For example, we can look at the nightlight satellite images of Barcelona from 1992 to 1997 based on the coordinates of the city that we can simply find on e.g. Google.

```python
plot_light_series('Barcelona_n41.39_e2.15', '1992-1997')
```

In the same fashion, we can also look at what is perhaps more interesting, the light changes between the years. To get e.g. the absolute change from year to year, be can use...

```python
plot_light_change('Barcelona_n41.39_e2.15', '1992-1997')
```

We can also look at the growth instead of the absolute change and over different periods of time as below...

```python
plot_light_growth('Barcelona_n41.39_e2.15', '1992_1993_2012_2013')
```

#### Styles
Whenever plotting the satellite data as an image, we can choose a style. The standard for displaying actual lights is usually `bone` and 'RdBu' for showing light differences, but there are other options. You can find the complete list of styles available through Matplotlib [here](https://matplotlib.org/examples/color/colormaps_reference.html).

```python
plot_light_series('Barcelona_n41.39_e2.15', '1993_2013', style='cubehelix')
```

#### Extraction Area Size
You can specify the size of the images that you're looking to extract with the `size` arguement. By default, a 250x250 pixel area around the image will cut out.

```python
plot_light_series('Barcelona_n41.39_e2.15', '1993_2013', style='cubehelix', size='500px_500px')
```

#### Gif Animations
You can also create a moving `gif` animation from a light grid series instead of plotting it. The animation is then saved and can be downloaded:

```python
animate_light_series('Barcelona_n41.39_e2.15', '1992-1997', style='cubehelix')
```
